### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Creación del diccionario Facturación para almacenar las ventas por tienda
Facturación = {}
Facturación['Tienda 1'] = tienda['Precio'].sum()
Facturación['Tienda 2'] = tienda2['Precio'].sum()
Facturación['Tienda 3'] = tienda3['Precio'].sum()
Facturación['Tienda 4'] = tienda4['Precio'].sum()

print('Facturación por tienda')
print(Facturación) #Presentación de la facturación

print('')
# Cantidad de registros por tienda
print('Registros por tienda')
print('Tienda 1 posee', len(tienda), 'registros')
print('Tienda 1 posee', len(tienda2), 'registros')
print('Tienda 1 posee', len(tienda3), 'registros')
print('Tienda 1 posee', len(tienda4), 'registros')
print('')

# Rango de fechas por tienda
print('Fecha de inicio y fin de los registros por tienda')
print('Tienda 1 inicia en ', tienda["Fecha de Compra"].min(), 'y finaliza el ', tienda["Fecha de Compra"].max())
print('Tienda 2 inicia en ', tienda2["Fecha de Compra"].min(), 'y finaliza el ', tienda2["Fecha de Compra"].max())
print('Tienda 3 inicia en ', tienda3["Fecha de Compra"].min(), 'y finaliza el ', tienda3["Fecha de Compra"].max())
print('Tienda 4 inicia en ', tienda4["Fecha de Compra"].min(), 'y finaliza el ', tienda4["Fecha de Compra"].max())

```

# 2. Ventas por categoría

```python
# En la lista Categorías se almacenan las categorías que se ofrecen en cada tienda
# Sí una tienda no ofrece una de las categorías, el monto de venta será 0

Categorias = []
for fila in range(len(tienda[:])):
  T2 = tienda[fila:fila + 1]
  categoria = T2["Categoría del Producto"].values[0]
  if categoria not in Categorias:
    Categorias.append(categoria)

for fila in range(len(tienda2[:])):
  T2 = tienda2[fila:fila + 1]
  categoria = T2["Categoría del Producto"].values[0]
  if categoria not in Categorias:
    Categorias.append(categoria)

for fila in range(len(tienda3[:])):
  T2 = tienda3[fila:fila + 1]
  categoria = T2["Categoría del Producto"].values[0]
  if categoria not in Categorias:
    Categorias.append(categoria)

for fila in range(len(tienda4[:])):
  T2 = tienda4[fila:fila + 1]
  categoria = T2["Categoría del Producto"].values[0]
  if categoria not in Categorias:
    Categorias.append(categoria)

# En este caso, todas las tiendas venden todas las categorías

#Categorias #Presentación de las categorías

# En el diccionario VentaCategoria_Tn se almacenan las ventas por categoría para cada tienda donde, Tn corresponde a:
# Tienda 1 = T1, Tienda 2 = T2, Tienda 3 = T3 y Tienda 4 = T4

# Tienda 1
VentaCategoria_T1 = {}
suma_total_cat = 0
for categoria in Categorias:
  suma_cat = 0
  for fila in range(len(tienda[:])):
    T2 = tienda[fila:fila + 1]
    if T2["Categoría del Producto"].values[0] == categoria:
      suma_cat += T2["Precio"].values[0]
      suma_total_cat += T2["Precio"].values[0]
  VentaCategoria_T1[categoria] = suma_cat

# Tienda 2
VentaCategoria_T2 = {}
suma_total_cat = 0
for categoria in Categorias:
  suma_cat = 0
  for fila in range(len(tienda2[:])):
    T2 = tienda2[fila:fila + 1]
    if T2["Categoría del Producto"].values[0] == categoria:
      suma_cat += T2["Precio"].values[0]
      suma_total_cat += T2["Precio"].values[0]
  VentaCategoria_T2[categoria] = suma_cat

# Tienda 3
VentaCategoria_T3 = {}
suma_total_cat = 0
for categoria in Categorias:
  suma_cat = 0
  for fila in range(len(tienda3[:])):
    T2 = tienda3[fila:fila + 1]
    if T2["Categoría del Producto"].values[0] == categoria:
      suma_cat += T2["Precio"].values[0]
      suma_total_cat += T2["Precio"].values[0]
  VentaCategoria_T3[categoria] = suma_cat

# Tienda 4
VentaCategoria_T4 = {}
suma_total_cat = 0
for categoria in Categorias:
  suma_cat = 0
  for fila in range(len(tienda4[:])):
    T2 = tienda4[fila:fila + 1]
    if T2["Categoría del Producto"].values[0] == categoria:
      suma_cat += T2["Precio"].values[0]
      suma_total_cat += T2["Precio"].values[0]
  VentaCategoria_T4[categoria] = suma_cat

# presentación de las ventas por categoría por tienda
print('Ventas por categoría de cada tienda')
print("Tienda 1", VentaCategoria_T1)
print("Tienda 2", VentaCategoria_T2)
print("Tienda 3", VentaCategoria_T3)
print("Tienda 4", VentaCategoria_T4)
```

# 3. Calificación promedio de la tienda


```python
# En el diccionario Calificación se almacenan los promedios por tienda
CalificaciónPromedio = {}
CalificaciónPromedio['Tienda 1'] = tienda['Calificación'].mean()
CalificaciónPromedio['Tienda 2'] = tienda2['Calificación'].mean()
CalificaciónPromedio['Tienda 3'] = tienda3['Calificación'].mean()
CalificaciónPromedio['Tienda 4'] = tienda4['Calificación'].mean()

print('Calificación promedio por tienda')
print(CalificaciónPromedio) # Presentación de la calificación promedio
```

# 4. Productos más y menos vendidos

```python
# En la lista Productos se almacenan los productos que se venden en las tiendas
# Sí una tienda no ofrece un de esto productos, el monto de venta será 0.

Productos = []
for fila in range(len(tienda[:])):
  T2 = tienda[fila:fila + 1]
  producto = T2["Producto"].values[0]
  if producto not in Productos:
    Productos.append(producto)

for fila in range(len(tienda2[:])):
  T2 = tienda2[fila:fila + 1]
  producto = T2["Producto"].values[0]
  if producto not in Productos:
    Productos.append(producto)

for fila in range(len(tienda3[:])):
  T2 = tienda3[fila:fila + 1]
  producto = T2["Producto"].values[0]
  if producto not in Productos:
    Productos.append(producto)

for fila in range(len(tienda4[:])):
  T2 = tienda4[fila:fila + 1]
  producto = T2["Producto"].values[0]
  if producto not in Productos:
    Productos.append(producto)

#Productos #Presentación de los productos
# En este caso, todas las tiendas venden todos los productos

# En el diccionario ProductosVendidos_Tn se almacena los artículos vendidos por tienda donde, Tn corresponde a:
# Tienda 1 = T1, Tienda 2 = T2, Tienda 3 = T3 y Tienda 4 = T4

# Tienda 1
ProductosVendidos_T1 = {}
suma_total_prod = 0
for producto in Productos:
  suma_prod = 0
  for fila in range(len(tienda[:])):
    T2 = tienda[fila:fila + 1]
    if T2["Producto"].values[0] == producto:
      suma_prod += 1
      suma_total_prod += 1
  ProductosVendidos_T1[producto] = suma_prod

# Tienda 2
ProductosVendidos_T2 = {}
suma_total_prod = 0
for producto in Productos:
  suma_prod = 0
  for fila in range(len(tienda2[:])):
    T2 = tienda2[fila:fila + 1]
    if T2["Producto"].values[0] == producto:
      suma_prod += 1
      suma_total_prod += 1
  ProductosVendidos_T2[producto] = suma_prod

# Tienda 3
ProductosVendidos_T3 = {}
suma_total_prod = 0
for producto in Productos:
  suma_prod = 0
  for fila in range(len(tienda3[:])):
    T2 = tienda3[fila:fila + 1]
    if T2["Producto"].values[0] == producto:
      suma_prod += 1
      suma_total_prod += 1
  ProductosVendidos_T3[producto] = suma_prod

# Tienda 4
ProductosVendidos_T4 = {}
suma_total_prod = 0
for producto in Productos:
  suma_prod = 0
  for fila in range(len(tienda4[:])):
    T2 = tienda4[fila:fila + 1]
    if T2["Producto"].values[0] == producto:
      suma_prod += 1
      suma_total_prod += 1
  ProductosVendidos_T4[producto] = suma_prod


# Ordenamiento por venta en forma descendente para cada tienda
ProductosVendidos_T1 = sorted(ProductosVendidos_T1.items(), key=lambda item: item[1], reverse=True)
ProductosVendidos_T2 = sorted(ProductosVendidos_T2.items(), key=lambda item: item[1], reverse=True)
ProductosVendidos_T3 = sorted(ProductosVendidos_T3.items(), key=lambda item: item[1], reverse=True)
ProductosVendidos_T4 = sorted(ProductosVendidos_T4.items(), key=lambda item: item[1], reverse=True)

# Extracción de los 10 productos más vendidos y los 10 productos menos vendidos por tienda
ProductosMásVendidos_T1 = ProductosVendidos_T1[:10]
ProductosMenosVendidos_T1 = ProductosVendidos_T1[-10:]

ProductosMásVendidos_T2 = ProductosVendidos_T2[:10]
ProductosMenosVendidos_T2 = ProductosVendidos_T2[-10:]

ProductosMásVendidos_T3 = ProductosVendidos_T3[:10]
ProductosMenosVendidos_T3 = ProductosVendidos_T3[-10:]

ProductosMásVendidos_T4 = ProductosVendidos_T4[:10]
ProductosMenosVendidos_T4 = ProductosVendidos_T4[-10:]


# Presentación de los resultados por tienda
print('Productos más vendidos y menos vendidos por tienda')
print("Tienda 1 - Productos más vendidos",ProductosMásVendidos_T1)
print("Tienda 1 - Productos menos vendidos",ProductosMenosVendidos_T1)
print("Tienda 2 - Productos más vendidos",ProductosMásVendidos_T2)
print("Tienda 2 - Productos menos vendidos",ProductosMenosVendidos_T2)
print("Tienda 3 - Productos más vendidos",ProductosMásVendidos_T3)
print("Tienda 3 - Productos menos vendidos",ProductosMenosVendidos_T3)
print("Tienda 4 - Productos más vendidos",ProductosMásVendidos_T4)
print("Tienda 4 - Productos menos vendidos",ProductosMenosVendidos_T4)
```

# 5. Envío promedio por tienda

```python
# En el diccionario CostoEnvíoPromedio se almacena el promedio de los costos de envío por tienda
# Se utiliza el método mean() parab obtener el promedio.
CostoEnvíoPromedio = {}
CostoEnvíoPromedio['Tienda 1'] = tienda['Costo de envío'].mean()
CostoEnvíoPromedio['Tienda 2'] = tienda2['Costo de envío'].mean()
CostoEnvíoPromedio['Tienda 3'] = tienda3['Costo de envío'].mean()
CostoEnvíoPromedio['Tienda 4'] = tienda4['Costo de envío'].mean()

# Presentación del resultado por tienda
print('Envío promedio por tienda')
print(CostoEnvíoPromedio)
```
